```python
df10 = pd.read_csv(r'C:\NIRSII\To_git\data\Aptamers\aptamers_from_aptagen.csv', index_col=0)
df10['Category'].value_counts()
```

```python
df_mol = df10[df10['Category'] == 'Small Organic']
df_mol
```

```python
import pubchempy as pcp
```

```python
for n in range(df10.shape[0]):
    name = df10.loc[n, 'Target']
    print(name)
    compounds = pcp.get_compounds(name, namespace='name')

    for compound in compounds:
        # Получение названия (IUPAC Name)
        smiles = compound.isomeric_smiles
        
        # Вывод названия
        print(smiles)
        df10.loc[n, 'smiles'] = smiles
```

```python
filtered_df = df10[df10['smiles'].notnull()]

print(filtered_df)
```

```python
filtered_df.to_csv(r'C:\NIRSII\To_git\data\Aptamers\processed\aptagen_small_molecules.csv')
```
